##### <font color = 'darkblue' > 5) Verificar a quantidade de valores nulos </font>

 <font color = 'darkblue' > O Princípio de Pareto (80/20) não é um teorema matemático e não se aplica diretamente à questão dos dados faltantes. Ele sugere que uma pequena porcentagem das causas (20%) é responsável por uma grande porcentagem dos efeitos (80%). Ou seja, se até 20% dos dados estiverem faltando pode preencher sem medo, acima disso, cai no viés. Exemplo: N= 53   Faltantes=30 30/53 = 56,60% </font>

<font color = 'darkblue' > Neste caso, temos metade do conjunto de dados com dados faltantes, vamos analisar por cada coluna se ultrapassa 20% de dados faltantes. 
Caso ultrapassa, vamos excluir os valores nulos fazer um drop na variável, caso contrário, vamos trata-la com a imputação da mediana. </font>

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np  # Importe numpy

# Carregar variáveis numéricas para plotagem
variaveis_numericas = []
for i in df.columns[0:13].tolist():
    if df.dtypes[i] == 'int64' or df.dtypes[i] == 'float64':            
        print(i, ':' , df.dtypes[i]) 
        variaveis_numericas.append(i)
  
plt.rcParams["figure.figsize"] = [10.00, 10.00]
plt.rcParams["figure.autolayout"] = True

plt.rcParams['font.size'] = 8  # Reduzi o tamanho geral da fonte
plt.rcParams['axes.titlesize'] = 12  # Reduzi o tamanho da fonte dos títulos

# Ajusta dinamicamente o número de linhas e colunas
n = len(variaveis_numericas)
linhas = int(np.ceil(n / 4))
colunas = 4

f, axes = plt.subplots(linhas, colunas, figsize=(15, 5*linhas))  # Ajuste o tamanho da figura para melhor visualização

linha = 0
coluna = 0

for i in variaveis_numericas:
    sns.histplot(data=df, x=i, kde=True, color="navy", ax=axes[linha, coluna])  # Histograma com KDE em azul marinho
    axes[linha, coluna].set_title(f'Distribuição de {i}')  # Adiciona título para cada subplot
    coluna += 1
    if coluna == 4:
        linha += 1
        coluna = 0

# Remove subplots vazios, se houver
while coluna < colunas and linha == linhas - 1:
    f.delaxes(axes[linha, coluna])
    coluna += 1

plt.tight_layout()  # Ajusta o layout para evitar sobreposição
plt.show()
```

<p> <font color= 'darkblue'> Em probabilidade e estatística, a correlação, dependência ou associação refere-se a qualquer relação estatística entre duas variáveis.

<p> O estudo da correlação tem como objetivo determinar e quantificar o grau de relacionamento entre essas variáveis (por exemplo, X e Y).

O valor do coeficiente de correlação varia entre -1 e 1, onde o resultado obtido indica se a correlação é negativa ou positiva.

- <font color='darkblue'> **Tratar Valores Faltantes:** Usamos a moda para preencher valores ausentes em variáveis categóricas e a mediana para variáveis numéricas.
- <font color='darkblue'> **Label Enconder:** Aplicamos nas variáveis que possui categorias ordinais "ordenadas" como (Numero_assentos, Numero_portas, Tamanho_pneu).

##### <font color= 'blue'> 6) Tratamento de Outliers  </font>

```python
import pandas as pd
from scipy import stats

# Supondo que você já tenha X_train e X_test definidos

# Identificar as features que precisam de transformação Box-Cox
skewed_features = ['Preço_minimo', 'Potencia_motor', 'Torque_maximo', 'Capacidade_bateria', 
                   'Alcance_WLTP', 'Distancia_eixos', 'Comprimento', 'Largura', 
                   'Altura', 'Peso_vazio_minimo', 'Peso_bruto_permitido', 
                   'Capacidade_maxima_carga', 'Velocidade_maxima', 
                   'Capacidade_porta_malas', 'Aceleracao', 'Potencia_maxima_carga']

# Função para aplicar Box-Cox
def apply_boxcox(df1, features):
    df_outlier = df1.copy()
    for feature in features:
        # Adicionar um valor constante para que todos os valores sejam positivos
        df_outlier[feature] = df_outlier[feature] + abs(df_outlier[feature].min()) + 1
        df_outlier[feature], _ = stats.boxcox(df_outlier[feature])
    return df_outlier

# Aplicar transformação Box-Cox em X_train
X_train = apply_boxcox(X_train, skewed_features)

# Aplicar transformação Box-Cox em X_test
X_test = apply_boxcox(X_test, skewed_features)
```

##### <font color= 'darkblue' > Treinamento, Validação e Teste </font>

```python
# Definindo a seed para reprodutibilidade
seed = 42

# Criação dos modelos com random_state definido para os que suportam
models = {
    'Ridge': Ridge(random_state=seed),
    'Lasso': Lasso(random_state=seed),
    'ElasticNet': ElasticNet(random_state=seed),
    'Decision Tree': DecisionTreeRegressor(random_state=seed),
    'Random Forest': RandomForestRegressor(n_estimators=100, random_state=seed),
    'Gradient Boosting': GradientBoostingRegressor(random_state=seed),
    'SVR': SVR(),
    'KNeighbors': KNeighborsRegressor(),
    'Linear Regression': LinearRegression()
}

# Dicionário para armazenar os resultados
results = {}

# Função para avaliar os modelos com validação cruzada usando RMSE e R²
def avaliar_modelo(model, X_train, y_train):
    # Validação cruzada com 5 folds para RMSE
    scores_rmse = cross_val_score(model, X_train, y_train, cv=5, scoring='neg_mean_squared_error')
    rmse_scores = np.sqrt(-scores_rmse)  # Calcula o RMSE para cada fold
    
    # Validação cruzada com 5 folds para R²
    scores_r2 = cross_val_score(model, X_train, y_train, cv=5, scoring='r2')
    
    return rmse_scores.tolist(), scores_r2.tolist()

# Avaliação dos modelos usando validação cruzada e dados de teste
for model_name, model in models.items():
    rmse_scores_cv, r2_scores_cv = avaliar_modelo(model, X_train, y_train)
    
    # Treinando o modelo com os dados de treino para avaliação no conjunto de teste
    model.fit(X_train, y_train)
    
    # Fazendo previsões nos dados de teste
    y_pred_test = model.predict(X_test)
    
    # Calculando RMSE e R² para os dados de teste
    rmse_test = np.sqrt(mean_squared_error(y_test, y_pred_test))
    r2_test = r2_score(y_test, y_pred_test)
    
    results[model_name] = {
        'RMSE Train (CV)': np.mean(rmse_scores_cv),  # Média do RMSE da validação cruzada
        'R² Train (CV)': np.mean(r2_scores_cv),      # Média do R² da validação cruzada
        'RMSE Test': rmse_test,
        'R² Test': r2_test
    }

# Convertendo os resultados em um DataFrame para melhor visualização
results_df = pd.DataFrame(results).T

# Exibindo os resultados
print(results_df)
```

```python
# Definindo a seed para reprodutibilidade
seed = 42

# Aplicando RFE para seleção de features usando um modelo base (por exemplo: Linear Regression)
base_model = LinearRegression()  # Modelo base para o RFE
n_features_to_select = 5  # Número de features desejadas (ajuste conforme necessário)
selector = RFE(estimator=base_model, n_features_to_select=n_features_to_select)

# Ajustando o seletor aos dados de treino e transformando os dados de treino e teste
X_train_selected = selector.fit_transform(X_train, y_train)
X_test_selected = selector.transform(X_test)

# Exibindo as features selecionadas (se os nomes das colunas estiverem disponíveis)
try:
    selected_features = X_train.columns[selector.support_]
    print("Features selecionadas:", selected_features.tolist())
except AttributeError:
    print("Não foi possível exibir os nomes das features selecionadas.")

# Criação dos modelos com random_state definido para os que suportam
models = {
    'Ridge': Ridge(random_state=seed),
    'Lasso': Lasso(random_state=seed),
    'ElasticNet': ElasticNet(random_state=seed),
    'Decision Tree': DecisionTreeRegressor(random_state=seed),
    'Random Forest': RandomForestRegressor(n_estimators=100, random_state=seed),
    'Gradient Boosting': GradientBoostingRegressor(random_state=seed),
    'SVR': SVR(),
    'KNeighbors': KNeighborsRegressor(),
    'Linear Regression': LinearRegression()
}

# Dicionário para armazenar os resultados
results = {}

# Função para avaliar os modelos com validação cruzada usando RMSE e R²
def avaliar_modelo(model, X_train, y_train):
    # Validação cruzada com 5 folds para RMSE
    scores_rmse = cross_val_score(model, X_train, y_train, cv=5, scoring='neg_mean_squared_error')
    rmse_scores = np.sqrt(-scores_rmse)  # Calcula o RMSE para cada fold
    
    # Validação cruzada com 5 folds para R²
    scores_r2 = cross_val_score(model, X_train, y_train, cv=5, scoring='r2')
    
    return rmse_scores.tolist(), scores_r2.tolist()

# Avaliação dos modelos usando validação cruzada e dados de teste
for model_name, model in models.items():
    rmse_scores_cv, r2_scores_cv = avaliar_modelo(model, X_train_selected, y_train)
    
    # Treinando o modelo com os dados de treino para avaliação no conjunto de teste
    model.fit(X_train_selected, y_train)
    
    # Fazendo previsões nos dados de teste
    y_pred_test = model.predict(X_test_selected)
    
    # Calculando RMSE e R² para os dados de teste
    rmse_test = np.sqrt(mean_squared_error(y_test, y_pred_test))
    r2_test = r2_score(y_test, y_pred_test)
    
    results[model_name] = {
        'RMSE Train (CV)': np.mean(rmse_scores_cv),  # Média do RMSE da validação cruzada
        'R² Train (CV)': np.mean(r2_scores_cv),      # Média do R² da validação cruzada
        'RMSE Test': rmse_test,
        'R² Test': r2_test
    }

# Convertendo os resultados em um DataFrame para melhor visualização
results_df = pd.DataFrame(results).T

# Exibindo os resultados
print(results_df)
```
